We would like to visualize data for specific points in time a well as across the past decade of data. To do this you wil lbe merging data from 2010 to 2022. Like you practiced in the control structures chapter, you should load in the multi-year ACS data with a for-loop, and a list of file names drawn from the files in the data folder. First make the vectors that the loop will iterate over. 

In this case it is the list of CSV files which we searched for with the argument `if i.endswith('Data.csv)`. The search string can't be just the .csv file extension, because the ACS data dictionaries are also CSV files. You also need a vector of just the years in question, because we will be adding the year as a value in a new column for each loaded dataframe.  

```python
path = '../../Data/ACS/DP04/Place/'
csv_files = [i for i in os.listdir(path) if i.endswith('Data.csv')]
csv_files.sort() # need csv_files in chronological order
csv_files
years.remove(2020) # ACS is missing 1-year data in 2020
years
```

```python
housing=[]
housing=pd.DataFrame(housing) # ceate empty dataframe to take new data

for i in range(len(csv_files)):
    file_path=path+csv_files[i]
    temp=pd.read_csv(file_path, low_memory=False) # load unique dataframes to a temporary df
    temp['year']=years[i] # add a year variable
    print(csv_files[i],'had',len(temp),'rows and',len(temp.columns),'columns')
    housing=pd.concat([housing,temp]) # concatenate the data into 'housing'
housing.shape
```
