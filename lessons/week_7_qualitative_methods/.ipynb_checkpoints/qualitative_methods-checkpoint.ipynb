## str.contains()

You may wish to search through your open-ended string variables for specific content, the way you would search inside of a document for a specific word. You have already done this in previous chapters with pandas' `str.contains()` function to subset by a specific country name. You can continue using str.contains() to look for observations with a specific word or phrase. Imagine you wished to analyze records of conflict involving the military, you could search for the terms 'navy', 'army' , 'air force', 'military' (and maybe 'police' if appropriate) in the narrative `notes` column using `str.contains()`. Note that we use the symbol `|`, which represents an `OR` condition, for _any_ of the previous terms, not _all_ the terms. The new subset dataframe called `military_conflicts` has 11,411 observations, down from the original 107,238 observations of the `south_america` dataframe. 

```python
%%time 
#check how long the query takes

military_conflicts = south_america[(south_america['notes'].str.contains('army|navy|air force|military', case=False))]

military_conflicts['fatalities'].describe()
```

We could also use a regular expression pattern to find the query in less lines of code. With regular expressions, you can ignore upper and lowercase with a global case-insensitive argument `(?i)`, followed by the four terms for armed forces we used above. Each term is separated by the `|` symbol again, and for 'air force' you will need to escape the white space using `\s` instead of a blank space.

```python
%%time 

#check how long the query takes
military_conflicts = south_america[south_america['notes'].str.contains(r'(?i)army|navy|air\sforce|military')]

military_conflicts['fatalities'].describe()
```

The code is very similar although we won't claim its much simpler. It comes down to your comfort and familiarity with regex. We included a `%%time` command at the start of each cell so you could observe any difference in time between commands. Regardless of your preference for a query, there will be times when using regex is unavoidable; when the text you need to extract are patterns and not exact word matches. 

## str.extract()

The event_id_cnty variable holds the three letter ISO country code plus the five number unique event id. The pattern is very simple to identify: Three letters from the alphabet, followed by five digits.Consider the first observation for cntry_event_id: 

`COL10767`

We could use any of these patterns to extract the even and ISO. 

- `^.{3}` The first three symbols regardless whether letters or numbers gets country ISO. 
- `[A-Z]+` One or more uppercase letters gets country ISO.
- `[A-Z]{3}` Exactly three uppercase letters gets country ISO.
- `.{5}$` The last five symbols gets event ID.
- `\d+` One or more numbers gets event ID.
- `\d{5}` Exactly five numbers gets event ID. 
- `[^\d]{3}` Three symbols that are __not__ numbers gets country ISO.
- `[^A-Z]{5}` Five symbols that are __not__ letters gets event ID.

try a few of these expressions out. Careful, however, to place the pattern to extract within parentheses. The str.contains() and str.match() functions don't care, but str.extract() will pull a pattern, and thus expects a distinguishing patter group to pull.

```python
# use str_extract()
south_america['event_id_cnty'].str.extract(r'(^.{3})') 
```

In the code cell below you can now create two new columns with extract and insert. You could also use str.split(), which you have used before, and provide a pattern as the delimiter. For now create a vector for either event id and ISO codes with str.extract(). It is important in this case to use the `expand=False` argument to obtain a flat vector to insert. By default the str.extract() function will output a grouped table, which doesn't play well with the insert() function.

```python
country_iso = south_america['event_id_cnty'].str.extract(r'([A-Z]{3})', expand=False)
event_id = south_america['event_id_cnty'].str.extract(r'(\d{5})', expand=False)

south_america.insert(1, 'country iso', country_iso)
south_america.insert(2, 'event id', event_id)
south_america.drop(columns='event_id_cnty', inplace=True)

south_america.head()
```
