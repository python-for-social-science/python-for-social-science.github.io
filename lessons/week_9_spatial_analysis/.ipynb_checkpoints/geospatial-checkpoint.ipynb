# Geospatial Analysis


## Background: Systems for Measuring Location 

### Latitude and Longitude
Latitude and longitude are a pair of coordinates used to specify locations on the Earth's surface. Latitude measures the north-south position on the Earth. It is the angle between a point on the Earth's surface and the equatorial plane, ranging from 0° at the Equator to 90° at the poles. Latitudinal lines, known as parallels, run east-west around the globe.  Longitude measures the east-west position relative to a prime meridian. It is the angle east or west from a reference line (meridian) that runs from the North Pole to the South Pole. In practice, the meridian used as a reference for longitude is the Prime Meridian; it is the planet's line of zero longitude. In 1884, the International Meridian Conference adopted the meridian passing through the Royal Observatory in Greenwich, London, as the universal Prime Meridian. This line divides the Earth into the Eastern and Western Hemispheres.


The system of latitude and longitude as we know it today evolved over many centuries. The ancient Greeks made significant contributions, with figures like Hipparchus credited with establishing the first known system of latitude and longitude. One of the most notable figures in this achievement was Eratosthenes in the 3rd century BCE. He is credited with being one of the first to calculate the Earth's circumference with reasonable accuracy. Eratosthenes used the angles of shadows cast by the sun at two different locations at the same time to estimate the Earth's circumference, which in turn provided a foundation for accurately determining latitude.

Longitude determination was a significant problem for navigators until the 18th century. The challenge was to accurately determine the time difference between a given location and a reference point (such as the Greenwich Meridian). Since the Earth rotates at a steady rate of 15 degrees per hour, knowing the time difference allows for the calculation of longitude.

John Harrison, an English carpenter and clockmaker, developed the marine chronometer, a timekeeping device precise enough to be used on sea voyages to determine longitude by means of celestial navigation. Prior to Harrison's work, there was no reliable method for determining longitude at sea. In 1735, he presented his first reliable chronometer, and his subsequent designs significantly improved long-distance sea travel by making navigation much more accurate and safe. It was one of the key technological advancements that enabled the Age of Discovery and the expansion of European colonialism across the oceans.

There are several systems for measuring longitude and latitude.  The best known is the Global Positioning System, GPS.  GPS was developed by the United States Department of Defense in the 1970s. It was initially intended for military applications, but in the 1980s, it was made available for civilian use. The system became fully operational in 1995 and has since revolutionized navigation and positioning services worldwide.

Other governments have developed their own global navigation satellite systems. The Russian GLONASS, the European Union's Galileo, and China's BeiDou are notable examples. These systems function similarly to GPS, offering global coverage and varying degrees of accuracy and reliability.

### Geohashes
Geohashes are a system of encoding geographic locations into a short string of letters and digits. They are hierarchical, with each additional character providing a more precise location. This system divides the Earth into a grid and assigns each cell a unique code.

#### Geohash Resolution Table

| Characters | Approximate Resolution |
|:----------:|:----------------------:|
| 2          | ±1100 km               |
| 3          | ±130 km                |
| 4          | ±30 km                 |
| 5          | ±2.4 km                |
| 6          | ±0.61 km               |
| 7          | ±0.076 km              |
| 8          | ±0.019 km              |
| 9          | ±0.0024 km             |


### what3words
what3words divides the world into a grid of 3m x 3m squares, each assigned a unique combination of three words. This system makes it easier to communicate precise locations. Unlike traditional GPS coordinates or geohashes, what3words addresses are easy to remember and share.

# Working with Geospatial Data
The most common geospatial formats is the shapefile, and libraries such as GDAL/OGR, Fiona, Pyogrio, and geopandas form the backbone of spatial data manipulation in Python. 

A shapefile is a popular geospatial vector data format in geospatial analysis. It is widely used for storing the location, shape, and attributes of geographic features, such as points, lines, and polygons. A shapefile is actually a collection of at least three files:

1. .shp: Contains the geometry of points, lines, or polygons.
2. .shx: Index file for the geometry.
3. .dbf: Contains attribute information in a dBASE format.

Despite its name, a shapefile is not a single file but a combination of these files working together. Shapefiles are advantageous due to their simplicity and wide acceptance in various spatial analysis program.  

The Geospatial Data Abstraction Library (GDAL) and its sister library, OpenGIS Simple Features Reference Implementation (OGR,part of GDAL), are open-source libraries for reading and writing raster and vector geospatial data formats, respectively. GDAL supports numerous raster formats, while OGR deals with vector formats like shapefiles.  They provide powerful and flexible tools for geospatial data conversion and processing such as reading and writing a wide range of spatial data formats, transforming geospatial data between different coordinate systems, accessing and manipulating raster data and metadata, and performing vector operations like feature selection and geometry transformations.

These pieces of software are not part of Python, and their appliation programming interfaces (APIs, how one accesses their capabilities) can be complex and less intuitive for beginners or those accustomed to Pythonic conventions.  In Python, GDAL and OGR are implemented through bindings that allow Python programs to use these libraries. 

Fiona is built on top of GDAL and OGR but with a more Pythonic interface. It focuses on making it easy to read and write geospatial data files with clear and concise code. Fiona adheres to Python standards and protocols, such as the context manager for opening and closing files. Key features of Fiona include simplified handling of geospatial data compared to raw GDAL/OGR bindings, reading and writing geospatial data in a dictionary-like format, and support for Pythonic idioms, making code more readable and maintainable.  Fiona is useful for scripts and applications that need direct interaction with geospatial data formats but does not provide many analytical capabilities.

Pyogrio is a newer library, providing a high-performance, Pythonic interface for reading and writing geospatial vector data. It is also built on top of GDAL and OGR but aims to simplify and quicken read and write operations for geospatial vector data, especially for large datasets.  Pyogrio is notable for fast operations for geospatial data and direct integration with Pandas.

This background is useful to know in order to appreciate the value of the `geopandas` library. GeoPandas is a high-level library that builds upon `pandas`,  `fiona`, and other dependencies, offering a more intuitive and powerful way of working with geospatial data.  GeoPandas introduces the `GeoDataFrame`, which extends the datatypes used by pandas to allow spatial operations on geometric types. Via its integration with `shapely`, another spatial analysis library, spatial operations such as spatial joins, overlays, and geocoding are possible.  It also provides convenient plotting and visualization capabilities thanks to Matplotlib. GeoPandas represents a significant advance for the open source analysis of spatial data.

### Making Maps
Outside of research settings, spatial analysis and map making are synonymous.  Probably because vision predates numeracy, most humans learn better from visuals of data than from looking directly at data.  For spatial data, that means it is more effective to plot variables on a map then to show descriptive statistics of those variables.  In addition, plotting on a map often requires less space than showing raw data or a description of those data.  For example, a map of the United States where each of its 3,143 counties is colored by population requires less space than a table with 3,143 rows.

This lesson therefore starts with plotting data before introducing more complicated programming.  The code below loads a shapefile of California boundaries.  It then shows different components to vary before coloring the counties by a variable.  Note that the variable, the county Federal Information Processing System (FIPS) number, is not something that would normally be visualized.  It is used here for ease of demonstration, and later examples use socially relevant data. 

```python
# The shapefile
ca_counties = gpd.read_file('../../Data/California/shapefiles/CA_Counties/CA_Counties_TIGER2016.shp')

# First plot
ca_counties.plot()

# No color
ca_counties.boundary.plot()

# Make the lines narrower
ca_counties.boundary.plot(linewidth=1)

# Narrow lines, no plot area border
ca_counties.boundary.plot(linewidth=1).set_axis_off()
```

You can also use `matplotlib` style syntax if you prefer.

```python
import matplotlib.pyplot as plt

ax = ca_counties.boundary.plot(linewidth=1)
ax.set_axis_off()
```

None of the figures above or below contain titles.  While it is common to find examples online inserting titles, adding titles when inserting a figure into a document (report, paper, etc.) is better.  If you insert a figure title now, it will have a different font than the report in which it is eventually included.  If you decide to change the title name, you have to update your code.  If you use numbering in the title ("Figure 1"), you will have to change that numbering when figure order changes.  For these reasons, it is best to not insert figure titles programmatically. 


Making outlines of administrative boundaries is nice, but outlines do not tell us anything on their own.  Instead, it is more useful to fill the boundaries in with a color, where the color corresponds to values of a variable.  This coloring is the primary power of spatial analysis as it allows people to see easily how an outcome, such as exposure to pollutants or household income, varies across space.  These type of plots are called *choropleths*. 

The county shapefile does not contain socially relevant variables on its own.  Once we learn how to color geographic units, then we will color them by a useful variable.

Since a figure not saved cannot be shared, it is therefore not much better than a figure not created.  The below code therefore saves the figure out as well.

The next set of code adds a legend and modifies its position.  Notice that the code imports `matplotlib`.  Finally, the code saves the figure to file. The first figure leaves too much whitespace, so the final line of code adds the `bbox_inches` argument to fix that.

```python
# Choropleth map (fill the shapes with colors) 
ca_counties.plot(column='COUNTYFP').set_axis_off()

ca_counties.plot(column='COUNTYFP', legend=True).set_axis_off()

ca_counties.plot(column='COUNTYFP', legend=True, legend_kwds={'title':'FIPS'}).set_axis_off()

ca_counties.plot(column='COUNTYFP', legend=True, legend_kwds={'title': 'FIPS', 'bbox_to_anchor': (1.25, 1)}).set_axis_off()

ca_counties.plot(column='COUNTYFP', legend=True, legend_kwds={'title': 'FIPS', 'bbox_to_anchor': (1.85, 1), 'ncol':3}).set_axis_off()

ca_counties.plot(column='COUNTYFP', legend=True, legend_kwds={'title': 'FIPS', 'bbox_to_anchor': (1.85, 1), 'ncol':3}).set_axis_off()
plt.savefig('../../Data/California/figures/ca_counties_fig1.jpg', dpi=300, format='jpg')

ca_counties.plot(column='COUNTYFP', legend=True, legend_kwds={'title': 'FIPS', 'bbox_to_anchor': (1.85, 1), 'ncol':3}).set_axis_off()
plt.savefig('../../Data/California/figures/ca_counties_fig1_tight.jpg', dpi=300, format='jpg', bbox_inches='tight')
```

There are four approaches to organizing colors used in spatial plotting; these approaches are often called maps or schemes.  *Sequential* colors change the saturation of one color or colors similar in hue to represent changes in values; these color schemes should be used when the plotted variable has ordering, preferably when it is continuously valued.  *Diverging* colors use two colors that meet at a variable's middle value; the variable values area again ordered and usually continuous.  The middle value is unsaturated (white), and the two colors can be any two but are usually similar in hue.  Diverging colors are best when there is a critical value to emphasize in the approximate middle of a range of values, such as a mean or median.  *Cyclic* colors are unsaturated at the minimum and maximum values of a variable.  It is not commonly used.  Finally, *qualitative* approaches represent nominal variables, variables with discrete unordered values.  These variables are most common in survey work; common exmaples include gender or religion.

`GeoPandas` relies on `matplotlib` for these color mappings.  To see all the color mappings available to the `cmap` argument, visit [this matplot page](https://matplotlib.org/stable/users/explain/colors/colormaps.html).  Color choices using more complicated mappings between values and colors are explored more in the next section.

You may have noticed that the figure made so far has repeating colors.  To avoid that, choose a different color mapping.  The below code uses the "cividis" colormapping, a sequential color scheme.  You can choose any mapping you like, though it should be from the Sequential or Sequential2 colormaps `matplotlib` provides.  While the FIPS code is qualitative, there are too many for the Qualitative color maps.

```python
ca_counties.plot(column='COUNTYFP', cmap='cividis', legend=True, legend_kwds={'title': 'FIPS', 'bbox_to_anchor': (1.85, 1), 'ncol':3}).set_axis_off()
plt.savefig('../../Data/California/figures/ca_counties_fig1_tight_cividis.jpg', dpi=300, format='jpg', bbox_inches='tight')
```

Maps are often more readable when their legends are on the bottom.  It is easy to change the legend position using `orientation` argument.  Oddly, `matplotlib` does not accept horizontal legends for non-numeric data.  The below code shows the error message, converts the column to the `float` type, then plots.

```python
# Will fail
# "title" is used because "label" is not recognized as a keyword argument.
ca_counties.plot(column='COUNTYFP', legend=True, legend_kwds={"title": "FIPS", "orientation": "horizontal"}).set_axis_off() 

# See COUNTYFP is object, which means not numeric
ca_counties.dtypes  

# Have to convert to numeric data (float or int)
ca_counties['COUNTYFP'] = ca_counties['COUNTYFP'].astype(float)

# Works
ca_counties.plot(column='COUNTYFP', legend=True, legend_kwds={"label": "FIPS", "orientation": "horizontal"}).set_axis_off() 
```

Seeing there is not a coordinate system, we can add it two ways.  First, we will geocode based on the city name.  Notice that the column names in each dataframe differ, so the code uses different column names.  In addition, one has to choose which internet service to use for the geocoding.  By default, `geopy`, the package `geopandas` uses, chooses Photon.  Photon is free and based on [OpenStreetMap](https://www.openstreetmap.org/#map=4/38.01/-95.84), itself an open source mapping project.  The `crowding` geocoding is commented out because it takes about 75 minutes to return the geometries.  The lethargy is because the code uses an open source free service.  Using a for-profit geocoder such as GoogleV3, Bing, or What3Words will return results more quickly, in return for money.  Using those services also requires passing credentials, a topic out of scope for now.  `crowding` receives its geometries in two cell chunks by merging with a shapefile.  

There are two other items of note.  First, the below code imports the geocode method.  If you do not want to import it, replace `geocode` with `gpd.tools.geocode`.  `gpd` is used because that is the name I gave the library when I imported it; change it if you use a different name.  Second, it is best practice for all imports to occur at the start of a script so that someone can quickly understand what libraries are required.  We put the import here because it is pedagogically easier.

```python
from geopandas.tools import geocode

# Geocoding 'geoname' in crowding and 'jurisdiction' in element using 'photon' provider
# crowding['geometry'] = geocode(crowding['geoname'] + ' California', provider='photon').geometry  # Takes approximately 75 minutes to run.
element['geometry'] = geocode(element['Jurisdiction']+ ', California', provider='photon').geometry

# Below is alternative syntax that avoids the import geocode line
# crowding['geometry'] = gpd.tools.geocode(crowding['geoname'], provider='photon').geometry
# element['geometry'] = gpd.tools.geocode(element['jurisdiction'], providder='photo').geometry
```

Notice that the above code adds ", California" to each place's name.  This is done because otherwise city names could be confused with more common place names.  For example, "Albany" will get geocoded to the New York City while "Dublin" will to the Irish capital.  "Albany, California" and "Dublin, California" will return the proper locations.

The second way is to merge each dataframe with a shapefile.  The below code loads a shapefile of California cities and then merges with `crowding`.  Geocoding with a shapefile is much quicker than geocoding via a third-party service.  The speed difference is especially pronounced when internet access in unavailable, such as on an airplane.  (Airplanes are for disconnection, please recharge your brain and avoid in-flight internet.)  In those instances, third-party services are unreachable and a shapefile is the only way to add geometries.

There are two subtleties to note in the below code.  First, notice that the resulting object uses a new name.  You do not have to use a new name, but it is done here to allow comparison of the original and merged objects.  Second, notice the merging is via a place's name, not a spatial coordinate.  This merging is no different than the merging of dataframes you have already learned.  Merging on place names is quite common but also quite brittle, as it requires place names to be spelled exactly the same way.  For example, if the left dataset uses "New York City" but the right uses "New York" or "NYC", `.merge()` will not add the right dataframe's records to "New York City".  The below code works smoothly because both datasets uses Census place names.

```python
# Load the shapefile
ca_places = gpd.read_file('../../Data/California/shapefiles/ca-places-boundaries/CA_Places_TIGER2016.shp')

# Look at columns to know which to use to merge.
ca_places.columns

# Merge
crowding_shp = crowding.merge(ca_places, left_on='geoname', right_on='NAMELSAD')

# Checking original versus merged
crowding.shape
crowding_shp.shape
crowding_shp.head()
```

The last three lines above confirm that `crowding` has fewer columns but the same number of rows as `crowding_shp`, and looking at the first five rows shows the new columns.  Everything looks good.

Now, confirm that there is a coordinate system in each dataframe and that they match.  `GeoPandas` does not recognize a coordinate system in `crowding_shp` because it expects coordinates to be named `geometry`, so we have to tell it the coordinate column.  We then delete the two columns `geometry_x` and `geometry_y` because `GeoPandas` will only save a `GeoDataFrame` with one geometry column, which it identifies as any column containing "geometry".

```python
crowding_shp.crs  # No
element.crs

# Make a geometry column for the column that actually has a geometry
crowding_shp['geometry'] = crowding_shp['geometry_y']
crowding_shp = crowding_shp.set_geometry('geometry') 

crowding_shp.crs  # Works

# Drop superfluous columns
crowding_shp = crowding_shp.drop(columns=['geometry_x', 'geometry_y'])
```

## Advanced Topics

### Coordinate Systems
The earth is three dimensional but coordinate systems are two dimensional.  There are several ways to convert a point in three dimensions to one in two.  These different ways are called coordinate systems. We saw this briefly in the _Geocoding and Merging_ section but did not focus on it.  Here, we do.

The below code loads data on transit stops in California.  It shows there is no defined geometry but there are columns called `X` and `Y`.  It then creates a geometry, defines the coordinate system, and converts it to the most common coordinate system. 

The final line converts the `ca_counties`' coordinate system.  In fact its original coordinate system is EPSG:3857, just like `transit`'s original.  One could therefore merge them without having converted coordinate systems (though `transit` still requires the `geometry` column to be made).  However, EPSG:4326 is the most common.

```python
# Load, inspect data
transit = gpd.read_file('../../Data/California/transit/CA_Transit_Stops.csv')
transit.head() # No geometry, x and y are not EPSG 4326 units.

# Make the geometry column, assuming X is longitude and Y is latitude
transit['geometry'] = gpd.points_from_xy(transit['X'], transit['Y'])

# Assume the projection system is EPSG 3857, planar. 
# planar means meters from the Prime Meridian and equator respectively.
transit.crs = 'epsg:3857'

# Convert to degrees, the longitude and latitude that is most common.
transit = transit.to_crs('epsg:4326')

# Below loads the shapefile.  While it has a geometry, the coordinate system is still different.
# transit_shp = gpd.read_file('../../Data/California/transit/RR_CA_Transit_Stops.shp')

# Convert ca_counties' coordinate system
ca_counties = gpd.read_file('../../Data/California/shapefiles/CA_Counties/CA_Counties_TIGER2016.shp')
ca_counties4326 = ca_counties.to_crs('epsg:4326')
```

### Attribute and Spatial Merging
Merging spatial data occurs two ways.  The first is by merging on a column that represents a spatial unit, such a city or county name; this column is the attribute on which merging occurs.  This merging is the same type of merging you learned earlier in the course, and it was used in `crowding.merge(places_shp, left_on='PL', right_on='PL')` to add geometries to the `crowding` dataset.  We review it again briefly now.  

First, load the data we will need.  If you have not already done so, also make sure to import `GeoPandas` and `matplotlib`.

```python
element = gpd.read_file('../../Data/California/housingElement/housing-element-compliance_shapefile.shp')
crowding_shp = gpd.read_file('../../Data/California/housingCrowding/hci_housing_crowding_shapefile.shp')
```

```python
# Make names lowercase so they match.  Otherwise, merging will fail.
element['County'] = element['County'].str.lower()
ca_counties['NAME'] = ca_counties['NAME'].str.lower()

element_shp = element.merge(ca_counties, left_on='County', right_on='NAME')

# make only one geometry column
element_shp['geometry'] = element_shp['geometry_y']
element_shp = element_shp.drop(columns=['geometry_x', 'geometry_y'])

# redefine the dataframe as a GeoDataFrame
element_shp = gpd.GeoDataFrame(element_shp)
```

The second way is an actual spatial merge, which is when observations from two datasets are merged based on their spatial relationship to each other.  There are two spatial merge methods, `.sjoin()` (often called "binary predicate joins") and `.sjoin_nearest()`.

To learn spatial merging, we will use the transit data.

```python
# Use a different object name to check the merge
transit2 = transit.sjoin(ca_counties4326, how='left', predicate='within')

# Explore, confirm new columns
transit.shape
transit2.shape
transit2.head()
```

"Predicate" means what type of spatial relationship is required for the merge.  The possible values are:
- intersects: True if the boundary and interior of one object intersect in any way with those of the other.
- contains: True if one object’s geometry is wholly within the interior of the other object’s geometry.
- within: The opposite of contains. True if one object’s geometry lies entirely within the other.
- touches: True if the objects have at least one point in common, but their interiors do not intersect.
- crosses: True if the interior of one object intersects the interior of the other but does not contain it, typically applying to lines intersecting or passing through one another.
- overlaps: True if the geometries share some but not all points in common, and the intersection of their interiors has the same dimension as the geometries themselves.

For more detail, see the [`shapely` documentation for binary predicates](https://shapely.readthedocs.io/en/latest/manual.html#binary-predicates).  

Nearest spatial joins are as follows.  Note that they can be a source of confusion depending on your data's coordinate system.  `.sjoin_nearest()` assumes the coordinates are planar (two-dimensional, no curvature of the earth).  It takes a `max_distance` argument and expects units in meters, e.g. 5 means 5 meters.  If your data are spheroid (latitude and longitude), the earth's curvature means distance becomes less accurate the further away two points are.  In other words, two points could be merged when they in fact should not be because their planar distance is less than `max_distance` though their spherical distance would not be.  These are rare cases and highly unlikely to occur in social science applications because merges occur between points very near each other, such as parks with bus stop data or parks with city data. 

Notice that the "geometry" column is a point, meaning it is from the `transit` dataframe.  The county polygon geometry was not merged in.  To do that now requires an attribute join, the traditional merge seen earlier in this lesson.  Fortunately, the "index_right" column was added.  A new object called `transit3` is created simply to make comparison easier.  Once you get comfortable with coding, it is not necessary to create new objects.

```python
transit3 = transit2.merge(ca_counties4326[['geometry']], left_on='index_right', right_index=True, how='left')

transit2.shape
transit3.shape
transit3.head()

# There are multiople geometries, fix.
# Below performs the same transformation as for element_shp, just with different syntax.
transit3.drop(columns=['geometry_x'], inplace=True)
transit3.rename(columns={'geometry_y': 'geometry'}, inplace=True)

# Convert back into GeoDataFrame
isinstance(transit3, gpd.GeoDataFrame)  
transit3 = gpd.GeoDataFrame(transit3)
```

### Aggregation
`element_shp` will not plot correctly because there are multiple observations per county.  Only one row per shape to be plotted is allowed, otherwise `GeoPandas` does not know which row to use.  It is therefore time to learn how to aggregate spatial data.  GeoPandas calls this procedure `dissolve()`, but it follows the same `group_by()` and `apply()` you have already learned.  We will take the `crowding` data and aggregate to the county level.  `GeoPandas` uses `pandas` for aggregation.  To see the default aggregation functions, visit the [`pandas` groupby documenation](https://pandas.pydata.org/pandas-docs/stable/user_guide/groupby.html#groupby-aggregate).  

Note that the below code aggregates on the `NAME` column.  It is better to use the county Federal Information Processing System (FIPS) code because names often vary across datasets -- "Los Angeles" versus "LA" for example -- whereas the county FIPS code will not.  Names are acceptable for this example because they both derive from the US Census.

In addition, notice that the below code creates a new column of numeric values.  This is necessary for aggregation to be able to make a quantitative analysis of a qualitative variable.

```python
# Make a new variable to aggregate
element_shp['Compliance_num'] = element_shp['Compliance'].apply(lambda x: 1 if x == 'IN' else 0)

# Check work
element_shp['Compliance_num'].value_counts()

# The below code is more readable but requires you to import numpy first.
#element_shp['Compliance_num'] = np.where(element_shp['Compliance'] == 'IN', 1, 0)

# Aggregate on "county_name", keeping the first value per county.
# as_index keeps the by column as a column.  Otherwise, it is transformed into an index.
agg_one = element_shp.dissolve(by='NAME', as_index=False)

# The above code fails because a DataFrame object has no dissolve attribute.
element_shp = gpd.GeoDataFrame(element_shp, geometry='geometry')

# Aggregation will work.  Inspect after.
es_one = element_shp.dissolve(by='NAME', as_index=False)
es_one.shape
es_one.head()

# Aggregate but taking the average and total number of cities in compliance by county
element_shp['Compliance_num2'] = element_shp['Compliance_num']
isinstance(element_shp, gpd.GeoDataFrame)  # False
element_shp = gpd.GeoDataFrame(element_shp)
es_two = element_shp.dissolve(by='NAME', aggfunc={'Compliance_num':'mean', 'Compliance_num2':'sum', 'ALAND': 'mean', 'AWATER': 'mean'}, as_index=False)
```

Now, aggregate the transit data to the county level.  There are 53,007 transit stops and 124,322 rows in `transit3`, so aggregating takes too long for teaching purposes.  The code below therefore first takes a sample of the data for plotting.  So long as your code uses the same value for `random_state`, you will get the same subset of data.  The aggregation is slow because the geometry column contains multipolygons, a "shape" that is multiple disconnected shapes.  Simplifying the polygon will accelerate the aggregation; documentation on how to do that [is available here](https://geopandas.org/en/stable/docs/reference/api/geopandas.GeoSeries.simplify.html#geopandas.GeoSeries.simplify) and otherwise outside the scope of this material.

The below code shows how to make the legend the same width as the plot area.  While not strictly necessary, it looks awkward to have a legend wider or taller than the plot itself.  There are two quirks about plotting horizontal legends to be aware of.  First, the horizontal legend is plotted to be the same width as the figure window regardless of the width of the map itself.  As a result, the legend is often much wider than the map, generating a lot of white space.  Second, `GeoPandas` does not actually create a legend, it creates a colorbar.  Keywords that `matplotlib` accepts for legends therefore may not work, which is why the below code makes the legend fit by adjusting the subplot size and not using the more common `bbox_to_anchor` argument.  

```python
# Create a figure and axes with adjusted size
fig, ax = plt.subplots(figsize=(4, 6))  # Adjust the first value, the width, to get a good fit.

# Plot the GeoDataFrame with the created axes object
es_two.plot(column='Compliance_num', ax=ax, legend=True, 
            legend_kwds={'label': 'Percent of Cities in Compliance', 
                         'orientation': 'horizontal'}).set_axis_off()
# Adjust layout to make room for the legend or other elements
fig.subplots_adjust(bottom=0.2, right=0.8)  # Adjust these values as needed
# Show the plot
plt.show()
```

One more powerful mapping capability is to change the measurements used to change the colors of a map.  For example, you may want to map income by country and divide counties into deciles such that the 10% of poorest counties use one color, the next 10% use a different one, and so on.  Or you may want to assign colors so that each color is applied to differentially sized intervals so that each interval contains an equal number of values regarldess of the range of values in that color.  Whereas percentiles force equal spacing on the variable values, the interval approach allows unequal spacing on the variable values so that there are an equal number of values per color.

`GeoPandas` relies on the `mapclassify` package for these more advanced approaches to color mapping.  `mapclassify` calls these mappings *schemes*, and they are called using the `scheme` argument of the `.plot()` method.  The [mapclassify API documentation](https://pysal.org/mapclassify/api.html) explains the 17 available schemes in more detail.  One of the 17 permits user defined mappings, meaning you as the analyst can create any mapping you want between a variable's values and its colors.

Finally, `mapclassify` is not installed as part of the Anaconda distribution or when `GeoPandas` is installed.  It is available via `conda`, the package manager Anaconda uses.  To install it, see *NOTE: INSERT REFERENCE TO LESSON WHERE WE TEACH HOW TO INSTALL PACKAGES*. 

```python
### Equal interval plotting
## Element compliance
es_two.plot(column='Compliance_num', 
            scheme='EqualInterval', 
            k=5, 
            legend=True).set_axis_off()

# Add a title
es_two.plot(column='Compliance_num', 
            scheme='EqualInterval', 
            k=5, 
            legend=True, 
            legend_kwds={'title': 'Percent of Cities in Compliance').set_axis_off()


## Crowding data
# First need to make the plotting column numeric.
# Then, subset.
crowding_shp['numerator'] = pd.to_numeric(crowding_shp['numerator'], errors='coerce')
cs_toplot = crowding_shp[crowding_shp['race_eth_n'] == 'Total']
cs_toplot = cs_toplot[cs_toplot['reportyear']=='2011-2015']

fig, ax = plt.subplots(figsize=(6, 10))
cs_toplot.plot(column='numerator', ax=ax, cmap='Oranges', 
              scheme='EqualInterval', k=5,  # This line uses mapclassify to create 5 intervals of the Oranges color mapping.
              legend=True, legend_kwds={'title': "Number of Overcrowded Households",
                                        'loc': 'upper right',
                                        }).set_axis_off()
plt.subplots_adjust(bottom=0.2)
plt.savefig('../../Data/California/figures/choropleth_numerator_orange_schemeEI.jpg', format='jpg')
plt.show()

# Plotting with pretty breaks
fig, ax = plt.subplots(figsize=(6, 10))
cs_toplot.plot(column='numerator', ax=ax, edgecolor='none', cmap='Oranges', 
              scheme='prettybreaks', k=5,  # This line uses mapclassify to create 5 intervals of the Oranges color mapping.
              legend=True, legend_kwds={'title': "Number of Overcrowded Households",
                                        'bbox_to_anchor': (1, .85)})
ax.axis('off') # This line turns the axes off
plt.subplots_adjust(bottom=0.2)
plt.savefig('../../Data/California/figures/choropleth_numerator_orange_schemePretty.jpg', format='jpg')
plt.show()
```

There are a few items worth noting in the above code.  First, `matplotlib` will plot `crowding_shp`, but that dataframe has multiple observations per place.  A subset of `crowding_shp` is therefore made so that each place only appears once.  Otherwise, every value of "Numerator" per place will be plotted but only the last one will appear. Second, the legend cannot easily be made horizontal because it is created from `mapclassify`.  Third, the "prettybreaks" scheme uses the `bbox_to_anchor` legend argument to show another method of changing the legend position.

### Subsetting by Bounding Boxes
Next, you may not want to plot all possible geometries in a dataframe.  For example, you may have a dataframe containing every county in California but only want those in the southern third.  One option is to select the desired counties by name, but that approach requires knowing the counties, and creating the object of county names is time consuming.  It would be even more difficult is one wants smaller geographic units like metropolitan statistical areas or Census tracts.  It is therefore often easiest to slice based on longitude and latitude.

The syntax is `object.cx[longitude_minimum:longitude_maximum, latitude_minimum:latitude_maximum]`.  Keep in mind that while longitude lines run vertically, their value changes as they west to east, so they represent the x values when plotting.  The opposite is true of latitude. 

To demonstrate this functionality, let us keep only those cities south of Santa Barbara, CA.  Since we are selecting on south, that means using latitude.

```python
# Load the data
crowding = gpd.read_file('../../Data/California/housingCrowding/hci_housing_crowding_shapefile.shp')

# Convert the coordinates from planar to degrees
crowding = crowding.to_crs('epsg:4326')

# Latitude of Santa Barbara, CA
santa_barbara_lat = 34.4208

# Filter the crowding GeoDataFrame using .cx indexer
# Usage: for longitude to select all longitudes, and specify latitude range
crowding_sb = crowding.cx[:, :santa_barbara_lat]

# Check
crowding_sb.shape
crowding.shape # More rows, good sign

crowding_sb.head()  # Encouraging
crowding_sb['county_nam'].value_counts()  # The expected counties!
```

Now one can mainpulate `crowding_sb` the same ways we have learned.  The below code does a very little bit of that.  The resulting plot has a lot of whtie space, which are missing cities.  To plot by county, aggregate.  To color the missing geometries, see [this GeoPandas documentation](https://geopandas.org/en/stable/docs/user_guide/mapping.html#missing-data).

```python
crowding_sb = crowding_sb[crowding_sb['race_eth_n'] == 'Total']
crowding_sb.plot(column='numerator').set_axis_off()
```
